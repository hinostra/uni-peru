## **Question 2**

¿Por qué los polinomios forman una estructura de anillo?

Porque cumple con los axiomas de estructura de anillo:

Sea $P$ el conjunto de todos los polinomios. Se dice que $P$ tiene estructura de anillo conmutativo pues en $P$ tenemos definidas dos operaciones, llamadas suma y producto, y denotadas por $+$ y $\cdot$, que satisfacen las siguientes propiedades:

* La suma es asociativa. Para cualesquiera $a, b, c \in P, a+(b+c) = (a+b)+c$
* La suma es conmutativa. Para cualesquiera $a, b \in P, a+b=b+a$.
* Elemento neutro de la suma. Existe un elemento $0 \in P$ tal que para cualquier $a \in P$ se verifica que $a+0=0+a=a$.
* Elemento opuesto. Dado $a \in P$, existe $b \in P$ tal que $a+b=0$. Este elemento, que es único, se denota por $-a$.
* El producto es asociativo. Para cualesquiera $a, b, c \in P, a \cdot (b \cdot c) = (a \cdot b)\cdot c$.
* El producto es conmutativo. Para cualesquiera $a, b \in P, a \cdot b = b \cdot a$.
* Elemento neutro del producto. Existe un elemento $1 \in P$ tal que para cualquier $a \in P$ se verifica que $a\cdot 1 = 1 \cdot a = a$.
* Distributividad. Para cualesquiera $a, b, c \in P$ se verifica que $a\cdot(b+c) =a\cdot b+a\cdot c$.


## **Question 3**

¿Qué es la Transformada de Fourier?

La transformada Fourier de una señal unidimensional o función continua $f(x)$ es una transformación de dicha señal que nos permite calcular la contribución de cada valor de frecuencia a la formación de la señal. La expresión matemática de dicho cálculo es:

$$F(u) = \int^{\infty}_{-\infty} f(x) \cdot e^{-i2\pi ux} dx$$

donde $i = \sqrt{-1}, e^{-i2\pi ux} = \cos (2\pi ux) - i \cdot \sin (2\pi ux)$ y la variable $u$ que aparece en la función $F(u)$ representa a las frecuencias.

## **Question 5**

Campo eléctrico

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.colors as colors
```

```python
q1, a1, b1 = 4.0e-6, -1.0, 1.0
q2, a2, b2 = 3.0e-6, -1.0, -1.0
q3, a3, b3 = -7.0e-6, 1.0, -1.0
```

**Campo Eléctrico**

```python
def E(x,y,a,b,q):
    Ex = q*(x-a)/((x-a)**2 + (y-b)**2)**(1.5)
    Ey = q*(y-b)/((x-a)**2 + (y-b)**2)**(1.5)
    
    return Ex, Ey

def modulo(Ax,Ay):
    A = (Ax**2 + Ay**2)**0.5
    
    return A
```

```python
xE = np.linspace(-5,5,10)
yE = np.linspace(-5,5,10)

XE, YE = np.meshgrid(xE,yE)
```

```python
Ex1, Ey1 = E(XE,YE,a1,b1,q1)
Ex2, Ey2 = E(XE,YE,a2,b2,q2)
Ex3, Ey3 = E(XE,YE,a3,b3,q3)

Etx = Ex1 + Ex2 + Ex3
Ety = Ey1 + Ey2 + Ex3

Et = modulo(Etx,Ety)

circle_q1 = plt.Circle((a1, b1), 0.15, color = 'r')
circle_q2 = plt.Circle((a2, b2), 0.15, color = 'r')
circle_q3 = plt.Circle((a3, b3), 0.15, color = 'k')

plt.quiver(XE, YE, Etx/Et, Ety/Et, Et, 
           pivot = 'middle', 
           cmap='spring', 
           norm = colors.LogNorm(vmin=Et.min(),vmax=Et.max()))
plt.gca().add_patch(circle_q1)
plt.gca().add_patch(circle_q2)
plt.gca().add_patch(circle_q3)
plt.axis('scaled')
plt.show()
```

**Líneas de Campo**

```python
plt.streamplot(XE, YE, Etx, Ety, 
               color = Et, 
               cmap='viridis', 
               arrowstyle='->')
plt.show()
```

**Potencial Eléctrico**

```python
def V(x, y, a, b, q):
    V = q/((x - a)**2 + (y - b)**2)**(1/2)
    
    return V
```

```python
xV = np.linspace(-5, 5)
yV = np.linspace(-5, 5)

XV, YV = np.meshgrid(xV, yV)

V1 = V(XV, YV, a1, b1, q1)
V2 = V(XV, YV, a2, b2, q2)
V3 = V(XV, YV, a3, b3, q3)

Vt = V1 + V2 + V3

plt.contourf(XV, YV, Vt, cmap = 'YlGnBu')
plt.contour(XV, YV, Vt, colors = 'k',linewidths=1)
plt.axis('scaled')
plt.show()
```
