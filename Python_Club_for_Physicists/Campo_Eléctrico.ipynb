```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.colors as colors
```

```python
q1, a1, b1 = 4.0e-6, -1.0, 1.0
q2, a2, b2 = 3.0e-6, -1.0, -1.0
q3, a3, b3 = -7.0e-6, 1.0, -1.0
```

# Campo El√©ctrico

```python
def E(x,y,a,b,q):
    Ex = q*(x-a)/((x-a)**2 + (y-b)**2)**(1.5)
    Ey = q*(y-b)/((x-a)**2 + (y-b)**2)**(1.5)
    
    return Ex, Ey

def modulo(Ax,Ay):
    A = (Ax**2 + Ay**2)**0.5
    
    return A
```

```python
xE = np.linspace(-5,5,10)
yE = np.linspace(-5,5,10)

XE, YE = np.meshgrid(xE,yE)
```

```python
Ex1, Ey1 = E(XE,YE,a1,b1,q1)
Ex2, Ey2 = E(XE,YE,a2,b2,q2)
Ex3, Ey3 = E(XE,YE,a3,b3,q3)

Etx = Ex1 + Ex2 + Ex3
Ety = Ey1 + Ey2 + Ex3

Et = modulo(Etx,Ety)

circle_q1 = plt.Circle((a1, b1), 0.15, color = '#DC143C')
circle_q2 = plt.Circle((a2, b2), 0.15, color = '#DC143C')
circle_q3 = plt.Circle((a3, b3), 0.15, color = '#1E90FF')

plt.quiver(XE, YE, Etx/Et, Ety/Et, Et, 
           pivot = 'middle', 
           cmap='spring', 
           norm = colors.LogNorm(vmin=Et.min(),vmax=Et.max()))
plt.gca().add_patch(circle_q1)
plt.gca().add_patch(circle_q2)
plt.gca().add_patch(circle_q3)
plt.axis('scaled')
plt.show()
```

# L√≠neas de Campo


```python
plt.streamplot(XE, YE, Etx, Ety, 
               color = Et, 
               cmap='viridis', 
               arrowstyle='->')
plt.show()
```

# Potencial El√©ctrico

```python
def V(x, y, a, b, q):
    V = q/((x - a)**2 + (y - b)**2)**(1/2)
    
    return V
```

```python
xV = np.linspace(-5, 5)
yV = np.linspace(-5, 5)

XV, YV = np.meshgrid(xV, yV)

V1 = V(XV, YV, a1, b1, q1)
V2 = V(XV, YV, a2, b2, q2)
V3 = V(XV, YV, a3, b3, q3)

Vt = V1 + V2 + V3

plt.contourf(XV, YV, Vt, cmap = 'hsv')
plt.contour(XV, YV, Vt, colors = 'k',linewidths=1)
plt.axis('scaled')
plt.show()
```
