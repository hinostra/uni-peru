```python
def plot3d_surfaces(ax,X,Y,Z,cmap,alpha=1,offset=None):
    surf = ax.plot_surface(X,Y,Z,cmap=cmap,rstride=1, cstride=1,alpha=alpha)
    ax.set_xlabel(r'$x \ axis$',fontsize=10,labelpad=1)
    ax.set_ylabel(r'$y \ axis$',fontsize=10,labelpad=1)
    ax.set_zlabel(r'$u \ axis$',fontsize=10,labelpad=5)
    ax.w_xaxis.set_pane_color((1.0, 1.0, 1.0, 1.0))
    ax.w_yaxis.set_pane_color((1.0, 1.0, 1.0, 1.0))
    ax.w_zaxis.set_pane_color((1.0, 1.0, 1.0, 1.0))
    #ax.contour(X, Y, Z, zdir='z', offset=offset, cmap=cmap,levels=[0])
    ax.tick_params(labelsize=10.0)
    #plt.xticks([0,0.79,1.57,2.36,3.14],['$0.0$','$\pi/4$','$\pi/2$','$3\pi/4$','$\pi$'],va='baseline')
    #plt.yticks([0,0.79,1.57,2.36,3.14],['$0.0$','$\pi/4$','$\pi/2$','$3\pi/4$','$\pi$'],va='baseline')
    plt.xticks([-5.0,-2.5,0.0,2.5,5.0],['$-5.0$','$-2.5$','$0.0$','$2.5$','$5.0$'],va='baseline')
    plt.yticks([-5.0,-2.5,0.0,2.5,5.0],['$-5.0$','$-2.5$','$0.0$','$2.5$','$5.0$'],va='baseline')
    #ax.set_zlim(0,0.2)
    plt.colorbar(surf,shrink=0.5) 
    ax.invert_xaxis()
    plt.show()
```

```python
mu = sol
r = np.linspace(R1, R2, 150)
p = np.linspace(0, 2*np.pi, 150)
R, P = np.meshgrid(r, p)
Z = 1.0*(-sc.yn(m,mu*R1)*sc.jv(m,mu*R)+sc.yn(m,mu*R)*sc.jv(m,mu*R1))*np.cos(m*P)
X, Y = R*np.cos(P), R*np.sin(P)
```

```python
fig = plt.figure(figsize=(1.2*6.4,1.2*4.8))
ax = axes3d.Axes3D(fig,rect=(0.0, 0.05, 0.95, 0.95),elev=35,azim=-60)
cmap = 'gist_earth'

plot3d_surfaces(ax,X,Y,Z,cmap,0.9,-0.065)
fig.savefig('anillosatanico23.pdf')
```
