# Coeficientes de Fresnel

```python
import numpy as np
import matplotlib.pyplot as plt
```

## Refracción externa

```python
n1 = 1.0
n2 = 1.5
n = n2/n1
```

Ángulo de incidencia $\theta_1$

```python
theta1 = np.linspace(0,np.pi/2,100)
theta = np.linspace(0,90,100)
theta2 = np.array([])
```

Coeficientes de reflexión

```python
for theta in theta1:
    theta2 = np.append(theta2,np.arcsin(np.sin(theta)/n))
```

```python
r_par1 = (n2*np.cos(theta1)-n1*np.cos(theta2))/(n2*np.cos(theta1)+n1*np.cos(theta2))
r_per1 = (n1*np.cos(theta1)-n2*np.cos(theta2))/(n1*np.cos(theta1)+n2*np.cos(theta2))
```

```python
theta.shape
```

```python
plt.plot(theta,r_par1,label='r_par')
plt.plot(theta,r_per1,label='r_perp')
plt.xlabel('Ángulo de incidencia')
plt.xlim([0,90])
plt.xticks([0,30,60,90],['0.0°','30.0°','60.0°','90.0°'])
plt.title('Coeficientes de Fresnel (aire-vidrio)')
plt.grid()
plt.legend()
plt.show()
```

```python
R_par1 = r_par1**2
R_per1 = r_per1**2
```

```python
plt.plot(theta,R_par1,label='R_par')
plt.plot(theta,R_per1,label='R_perp')
plt.xlabel('Ángulo de incidencia')
plt.xlim([0,90])
plt.xticks([0,30,60,90],['0.0°','30.0°','60.0°','90.0°'])
plt.title('Reflectancia (aire-vidrio)')
plt.grid()
plt.legend()
plt.show()
```
