```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def v(x,a):
    return ((1-np.cos(x))/(a*(1+np.cos(x))) + 0.5*np.cos(x))
```

```python
theta = np.linspace(-np.pi,np.pi,500)
a = np.array([15,0.15])
pot_est = v(theta,a[1])
pot_in = v(theta,a[0])
```

```python
fig, ax = plt.subplots(figsize=(8.3,6.2))
ax.plot(theta,pot_est,'b',label=r'V($\theta$) Estable')
ax.plot(theta,pot_in,'r',label=r'V($\theta$) Inestable')
ax.set_ylim([min(pot_in)-0.1,5])
ax.set_title('Potencial efectivo',fontsize=15)
ax.legend()
ax.set_xlabel(r'$\theta$',fontsize=15)
ax.set_ylabel(r'V($\theta$)',fontsize=15)
plt.savefig('Pot-efect-PC8.pdf')
plt.show()
```

```python
import sympy as sp
sp.init_printing()
```

```python
t = sp.symbols('t')
```

```python
R1 = sp.Matrix([[sp.cos(t),-sp.sin(t),0], 
               [sp.sin(t),sp.cos(t),0],
               [0,0,1]])

R2 = sp.Matrix([[sp.cos(t),0,sp.sin(t)], 
               [0,1,0],
               [-sp.sin(t),0,sp.cos(t)]])

R3 = sp.Matrix([[1,0,0],
               [0,sp.cos(t),-sp.sin(t)], 
               [0,sp.sin(t),sp.cos(t)]])
```

```python
R=R1*R2*R3
```

```python
sp.simplify(R**(-1))
```
