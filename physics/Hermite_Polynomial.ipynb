```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-2,2,250)
```

# Polinomios de Hermite

```python
def h0(x):
    return np.ones_like(x)

def h1(x):
    return 2*x
    
def h2(x):
    return 4*x**2-2

def h3(x):
    return 8*x**3-12*x

def h4(x):
    return 16*x**4-48*x**2+12
    
def h5(x):
    return 32*x**5-160*x**3+120*x
```

```python
fig, ax = plt.subplots(1,1,figsize=(10,8))
ax.plot(x,h0(x),'r',label=r'$H_0(x)$')
ax.plot(x,h1(x),'b',label=r'$H_1(x)$')
ax.plot(x,h2(x),'g',label=r'$H_2(x)$')
ax.legend()
fig.suptitle('Polinomios de Hermite $n=0,1,2$',fontsize=15,position = (0.5,0.92))
ax.grid()
plt.show()
fig.savefig('h0h1h2.pdf')
```

```python
figp, axp = plt.subplots(1,1,figsize=(10,8))
axp.plot(x,h3(x),'c',label=r'$H_3(x)$')
axp.plot(x,h4(x),'y',label=r'$H_4(x)$')
axp.plot(x,h5(x),'m',label=r'$H_5(x)$')
axp.legend()
axp.grid()
figp.suptitle('Polinomios de Hermite $n=3,4,5$',fontsize=15,position = (0.5,0.92))
plt.show()
figp.savefig('h3h4h5.pdf')
```

# Funciones de onda para el oscilador armónico cuántico

```python
from scipy.constants import *
from math import *
```

```python
a = (hbar/1.0e-39)**0.5
x1 = np.linspace(-1000,1000,2000)
```

```python
def Nn(n): 
    return 1/(a*(pi)**0.5*2**n*factorial(n))**0.5
```

```python
def fonda(n,x):
    return Nn(n)*np.exp(-x**2/(2*a**2))
```

```python
figfo, (ax2,ax3) = plt.subplots(1,2,figsize = (15,5))

ax2.plot(x1,h0(x1/a)*fonda(0,x1),'--',label='$\psi_1(x)$')
ax2.plot(x1,h1(x1/a)*fonda(1,x1),'-.',label='$\psi_2(x)$')
ax2.plot(x1,h2(x1/a)*fonda(2,x1),label='$\psi_3(x)$')
ax3.plot(x1,h3(x1/a)*fonda(3,x1),'r--',label='$\psi_4(x)$')
ax3.plot(x1,h4(x1/a)*fonda(4,x1),'k',label='$\psi_5(x)$')
ax3.plot(x1,h5(x1/a)*fonda(5,x1),'y-.',label='$\psi_6(x)$')
ax2.grid()
ax3.grid()
ax2.legend()
ax3.legend()
figfo.suptitle('Funciones de onda',fontsize=15,position = (0.5,0.94))
plt.show()

figfo.savefig('funcion_onda.pdf')
```

```python
figd, (ax4,ax5) = plt.subplots(1,2,figsize = (15,5))

ax4.plot(x1,(h0(x1/a)*fonda(0,x1))**2,'--',label='$|\psi_1|^2(x)$')
ax4.plot(x1,(h1(x1/a)*fonda(1,x1))**2,'-.',label='$|\psi_2|^2(x)$')
ax4.plot(x1,(h2(x1/a)*fonda(2,x1))**2,label='$|\psi_3|^2(x)$')
ax5.plot(x1,(h3(x1/a)*fonda(3,x1))**2,'r--',label='$|\psi_4|^2(x)$')
ax5.plot(x1,(h4(x1/a)*fonda(4,x1))**2,'k',label='$|\psi_5|^2(x)$')
ax5.plot(x1,(h5(x1/a)*fonda(5,x1))**2,'y-.',label='$|\psi_6|^2(x)$')
ax4.grid()
ax5.grid()
ax4.legend()
ax5.legend()
figd.suptitle('Distribución de probabilidades',fontsize=15,position = (0.5,0.94))
plt.show()

figd.savefig('distribuciones.pdf')
```
